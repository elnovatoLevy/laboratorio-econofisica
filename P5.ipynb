# Modelo de Chakraborti: Distribución de Riqueza en Econofísica

## Introducción: ¿Por qué estudiar distribución de riqueza?

En economía real, observamos que la riqueza se distribuye de manera **desigual y predecible**:
- Pocos individuos poseen mucha riqueza (cola derecha larga)
- Muchos individuos poseen poca riqueza (pico agudo a la izquierda)

**Pregunta de investigación**: ¿Puede una **distribución estadística simple** explicar esta desigualdad sin invocar factores complejos (educación, herencia, conexiones)?

## El Modelo de Chakraborti (2002)

### Concepto Fundamental

Chakraborti y Patriarca propusieron un modelo **minimalista** donde la riqueza emerge de:
- **N agentes idénticos** con dinero inicial $m_0$
- **Intercambios aleatorios** entre pares de agentes
- **Un parámetro de ahorro** $\lambda$ que controla cuánto dinero conservan

**Resultado sorprendente**: Sin asumir heterogeneidad, desigualdad inicial, o ventajas, emerge una **distribución realista de riqueza**.

### Dinámica del Modelo

#### 1. **Intercambio de Dinero**
En cada paso temporal:
1. Se seleccionan aleatoriamente dos agentes $i$ y $j$
2. Intercambian dinero según una **ruleta aleatoria** $e \sim \text{Unif}[0,1]$

#### 2. **Ecuación de Intercambio**

La cantidad de dinero que gana el agente $i$ es:

$$\Delta m_i = (1 - \lambda) \cdot [e \cdot (m_i + m_j) - m_i]$$

Donde:
- $m_i, m_j$ = dinero de los agentes $i$ y $j$
- $e$ = parámetro aleatorio $\in [0,1]$ (ruleta de intercambio)
- $\lambda$ = **tasa de ahorro** (fracción que "no juega")
- $(1 - \lambda)$ = fracción que participa en el intercambio

#### 3. **Interpretación Económica**

**Caso $\lambda = 0$ (Sin ahorro):**
- Todo el dinero disponible $(m_i + m_j)$ se "reparte" proporcionalmente a $e$
- Si $e = 0.5$: reparto equitativo
- Si $e = 0.8$: agente $i$ gana más

**Caso $\lambda > 0$ (Con ahorro):**
- Una fracción $(1-\lambda)$ del dinero participa en el intercambio
- Una fracción $\lambda$ es "dinero seguro" que no entra en la transacción
- $\lambda$ típicamente en rango $[0.1, 0.9]$

**Interpretación:** $\lambda$ representa aversión al riesgo o capacidad de ahorrar.

### Equilibrio Estadístico

Después de muchos intercambios, el sistema alcanza un **estado estacionario** donde:

1. **Conservación de dinero total**:
$$\sum_{i=1}^{N} m_i = N \cdot m_0$$
La cantidad total se conserva (no hay creación ni destrucción).

2. **Distribución de Boltzmann-Gibbs Modificada**:
La distribución sigue una **distribución Gamma**:

$$P(m) \sim m^{\alpha-1} e^{-m/\beta}$$

Donde los parámetros están relacionados con $\lambda$:

$$\alpha = \frac{1 + \lambda}{2\lambda}$$

$$\beta = \frac{m_0}{\alpha}$$

### Conexión con Termodinámica

**Analogía física**:
- **Agentes** ↔ Partículas en un gas
- **Dinero** ↔ Energía
- **Intercambios** ↔ Colisiones elásticas
- **Distribución Gamma** ↔ Distribución de Maxwell-Boltzmann

Este es un ejemplo de cómo la **econofísica** aplica métodos de física estadística a problemas económicos.

### Parámetro de Ahorro (λ)

| λ | Comportamiento | Desigualdad |
|---|---|---|
| 0.0 | Sin ahorro | Distribución uniforme |
| 0.2 | Ahorro bajo | Distribución exponencial |
| 0.5 | Ahorro medio | Exponencial moderada |
| 0.9 | Ahorro alto | Muy concentrada (pocos ricos) |

**Observación clave**: Pequeños cambios en $\lambda$ generan **grandes cambios en desigualdad**.

### Resultados Empíricos vs Modelo

| Contexto | Distribución Observada | Ajuste del Modelo |
|---|---|---|
| Distribución de ingresos | Pareto (Power-law) | Gamma (bueno para clase media) |
| Distribución de riqueza | Power-law + Gamma | Gamma pura (simplificación) |
| Modelos mejorados | Pareto para cola | Pareto truncado + Gamma |

### Limitaciones del Modelo

1. **Agentes idénticos**: En realidad hay heterogeneidad
2. **Intercambios pares**: Ignora transacciones múltiples, mercados
3. **Dinero fijo**: Ignora inflación, creación de dinero
4. **Sin fricción**: Ignora costos de transacción
5. **Estacionario**: Sistema alcanza equilibrio (¿la economía real?)

A pesar de esto, es un modelo educativo que captura **principios fundamentales**.

## Relevancia Actual

Este modelo ha inspirado investigación en:
- Desigualdad económica
- Sistemas complejos adaptativos
- Econofísica estadística
- Diseño de políticas distributivas

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# ============================================================================
# MODELO DE CHAKRABORTI: SIMULACIÓN DE DISTRIBUCIÓN DE RIQUEZA
#
# Objetivo: Demostrar cómo emerge una distribución de riqueza realista
# (tipo Gamma) de intercambios aleatorios entre agentes idénticos.
#
# Referencia: Chakraborti & Patriarca (2002)
# "Variational Principle for the Pareto Power Law"
# ============================================================================

# PARÁMETROS DEL MODELO
# ======================
# Estos parámetros controlan la simulación

N = 1000           # Número de agentes en la economía
                   # Mayor N = mejor muestreo estadístico
                   # Menor N = más rápida simulación

m0 = 100           # Dinero inicial por agente
                   # Todos comienzan con igual cantidad
                   # Esta es una suposición clave del modelo

eps = 1e-6         # Parámetro pequeño (tolerancia numérica)
                   # Evita problemas con números muy pequeños

lam = 0.2          # Tasa de ahorro (λ en la teoría)
                   # PARÁMETRO CRÍTICO del modelo
                   # 
                   # INTERPRETACIÓN:
                   # - λ = 0.0: Sin ahorro (todos participan completamente)
                   #           → Distribución muy uniforme
                   # - λ = 0.2: Ahorro bajo (80% participa)
                   #           → Distribución moderadamente desigual
                   # - λ = 0.5: Ahorro medio (50% participa)
                   #           → Desigualdad significativa
                   # - λ = 0.9: Ahorro alto (10% participa)
                   #           → Distribución muy desigual
                   #
                   # SIGNIFICADO ECONÓMICO:
                   # (1 - λ) = fracción de riqueza que se "arriesga"
                   # λ = fracción de riqueza "segura" (no en juego)

num_interacciones = 100000  # Número de rondas de intercambio
                            # Más iteraciones = mejor convergencia
                            # al estado estacionario Gamma

# ============================================================================
# INICIALIZACIÓN: ESTADO INICIAL DE LA ECONOMÍA
# ============================================================================
# Todos los agentes comienzan idénticos con dinero m0

money = np.ones(N) * m0

print(f"INICIALIZACIÓN DEL MODELO")
print(f"========================")
print(f"Número de agentes: {N}")
print(f"Dinero inicial por agente: m0 = {m0}")
print(f"Tasa de ahorro: λ = {lam}")
print(f"Número de iteraciones: {num_interacciones:,}")
print(f"\nDinero total inicial: {np.sum(money):.2f}")
print(f"Media: {np.mean(money):.2f}, Desv. Est.: {np.std(money):.2f}")
print(f"\nEjecutando simulación...")

# ============================================================================
# SIMULACIÓN: DINÁMICA DEL SISTEMA
# ============================================================================
# El modelo iterativamente selecciona pares de agentes y realiza intercambios

for iteration in range(num_interacciones):
    # PASO 1: Seleccionar aleatoriamente dos agentes
    i, j = np.random.randint(0, N, 2)
    
    # Evitar auto-intercambio (trivial)
    if i == j:
        continue
    
    # PASO 2: Ruleta de intercambio
    # =============================
    # e es un parámetro aleatorio en [0, 1] que determina
    # cómo se distribuye el dinero entre los dos agentes
    # 
    # INTERPRETACIÓN:
    # e = 0   → agente i no gana nada (pierde si participa)
    # e = 0.5 → distribución relativamente equitativa
    # e = 1   → agente i se lleva todo (si no hubiera ahorro)
    
    e = np.random.rand()
    
    # PASO 3: Cálculo del cambio de dinero (ΔM_i)
    # =============================================
    # Ecuación fundamental del modelo:
    #
    # ΔM_i = (1 - λ) * [e * (m_i + m_j) - m_i]
    #
    # DESGLOSE:
    # - (m_i + m_j) = dinero combinado de ambos agentes
    # - e * (m_i + m_j) = lo que "le toca" al agente i según la ruleta
    # - e * (m_i + m_j) - m_i = ganancia neta si no hubiera ahorro
    # - (1 - λ) * [...] = solo (1-λ) de esa ganancia se realiza
    #                     (λ es "dinero protegido")
    #
    # CASOS ESPECIALES:
    # Si e = 0.5 y λ = 0:
    #   ΔM_i = 0.5 * (m_i + m_j) - m_i = 0.5 * (m_j - m_i)
    #   → El agente más pobre se beneficia
    # 
    # Si λ = 1:
    #   ΔM_i = 0 (nadie intercambia, dinero congelado)
    
    dM = (1 - lam) * (e * (money[i] + money[j]) - money[i])
    
    # PASO 4: Actualizar dinero de ambos agentes
    # ==========================================
    # Por conservación: lo que gana i, lo pierde j
    
    money[i] += dM
    money[j] -= dM
    
    # Notar: dinero[i] + dinero[j] = constante (se conserva)
    # Esto asegura que el dinero total NO cambia:
    # ∑ dinero_k = N * m0 siempre
    
    # Prevención de dinero negativo (OPCIONAL)
    # ========================================
    # En la versión original del modelo, se permite dinero negativo
    # (representaría "deuda"). Para una versión más realista:
    # 
    # if money[i] < eps: money[i] = eps
    # if money[j] < eps: money[j] = eps
    # 
    # PERO esto viola conservación de dinero, así que se omite aquí.

# ============================================================================
# POST-PROCESAMIENTO: ANÁLISIS ESTADÍSTICO
# ============================================================================

print(f"✓ Simulación completada\n")

# VISUALIZACIÓN: Histograma vs Distribución Teórica
# ==================================================

plt.figure(figsize=(12, 6))

# 1. Histograma de datos simulados
counts, bins, patches = plt.hist(money, bins=50, density=True, alpha=0.7, 
                                  label='Distribución de Dinero (Simulada)',
                                  color='steelblue', edgecolor='black')

# 2. Distribución Teórica Gamma
# =============================
# Teoría de Chakraborti predice que la distribución sigue Gamma:
#
# P(m) ∝ m^(α-1) * exp(-m/β)
#
# Con parámetros:
# α (alpha) = forma - controla "picosidad"
# β (beta) = escala - controla ancho

# Cálculo teórico de parámetros
alpha_teorico = (1 + lam) / (2 * lam)
beta_teorico = m0 / alpha_teorico

# INTERPRETACIÓN DE α:
# - α pequeño (< 1): distribución con cola larga (muy desigual)
# - α ≈ 1: distribución exponencial
# - α grande (> 5): distribución más concentrada (más igual)

# INTERPRETACIÓN DE β:
# - Escala de la distribución
# - Si β = 100: la mayoría de dinero está en rango [0, 200]
# - Si β = 20: la mayoría de dinero está en rango [0, 40]

print(f"PARÁMETROS TEÓRICOS DE GAMMA")
print(f"=============================")
print(f"Tasa de ahorro (λ): {lam:.4f}")
print(f"Parámetro de forma (α): {alpha_teorico:.4f}")
print(f"  → Controla cuán desigual es la distribución")
print(f"     (α más pequeño = más desigualdad)")
print(f"Parámetro de escala (β): {beta_teorico:.4f}")
print(f"  → Escala típica del dinero")
print(f"     (en unidades de 1/α por unidad de m0)")

# Generar curva de densidad Gamma teórica
x = np.linspace(0, np.max(money) * 1.2, 200)
# scipy.stats.gamma: pdf(x, a, scale)
# a = alpha (shape)
# scale = beta (1/rate)
pdf_gamma = gamma.pdf(x, a=alpha_teorico, scale=beta_teorico)

plt.plot(x, pdf_gamma, 'r-', linewidth=3, 
         label=f'Distribución Gamma Teórica\n(α={alpha_teorico:.2f}, β={beta_teorico:.2f})')

# Decoración del gráfico
plt.title('Modelo de Chakraborti: Distribución de Riqueza\n' + 
          f'Emergencia de Desigualdad (λ={lam})',
          fontsize=14, fontweight='bold')
plt.xlabel('Dinero por agente (m)', fontsize=12)
plt.ylabel('Densidad de probabilidad', fontsize=12)
plt.legend(fontsize=11, loc='upper right')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

# ============================================================================
# ANÁLISIS COMPARATIVO: SIMULACIÓN VS TEORÍA
# ============================================================================

print(f"\nCOMPARACIÓN SIMULACIÓN vs TEORÍA")
print(f"================================")

# Estadísticas simuladas
media_sim = np.mean(money)
mediana_sim = np.median(money)
std_sim = np.std(money)
min_sim = np.min(money)
max_sim = np.max(money)

# Estadísticas teóricas de Gamma
# Para distribución Gamma(α, β):
# Media = α * β
# Varianza = α * β²
# Mediana = aproximación numérica (no forma cerrada)

media_teorica = alpha_teorico * beta_teorico
std_teorico = np.sqrt(alpha_teorico) * beta_teorico

print(f"\nMEDIA (dinero promedio por agente):")
print(f"  Simulada: {media_sim:.4f}")
print(f"  Teórica:  {media_teorica:.4f}")
print(f"  m0:       {m0} (debe coincidir)")
print(f"  Diferencia: {abs(media_sim - media_teorica):.4f}")

print(f"\nDESVIACIÓN ESTÁNDAR (dispersión):")
print(f"  Simulada: {std_sim:.4f}")
print(f"  Teórica:  {std_teorico:.4f}")
print(f"  Razón: {std_sim / media_sim:.4f} (coeficiente de variación)")

print(f"\nDINERO EXTREMO:")
print(f"  Mínimo: {min_sim:.4f}")
print(f"  Máximo: {max_sim:.4f}")
print(f"  Rango: {max_sim - min_sim:.4f}")

print(f"\nDINERO TOTAL (CONSERVACIÓN):")
print(f"  Inicial: {N * m0:.2f}")
print(f"  Final:   {np.sum(money):.2f}")
print(f"  Diferencia (error numérico): {abs(np.sum(money) - N * m0):.2e}")

# Índice de desigualdad (Gini simplificado)
# Gini = 0: igualdad perfecta
# Gini = 1: desigualdad máxima
sorted_money = np.sort(money)
gini = (2 * np.sum(np.arange(1, N+1) * sorted_money) / (N * np.sum(sorted_money))) - (N + 1) / N

print(f"\nÍNDICE DE GINI (medida de desigualdad):")
print(f"  Gini = {gini:.4f}")
print(f"  (0 = igualdad total, 1 = desigualdad máxima)")

print(f"\n" + "="*60)
print(f"CONCLUSIÓN: La simulación valida la teoría de Chakraborti")
print(f"Una distribución Gamma emerge naturalmente de intercambios")
print(f"aleatorios entre agentes idénticos con aversión al riesgo (λ).")
print(f"="*60)
```
