# Teorema del Límite Central (TLC) en Econofísica

## Introducción: El Fenómeno Universal

El **Teorema del Límite Central** es uno de los resultados más importantes en probabilidad y estadística. Explica por qué:

- Las distribuciones normales aparecen **ubícuamente en la naturaleza**
- Los errores de medición tienden a ser gaussianos
- Los retornos agregados de mercados convergen a la normalidad
- Muchos fenómenos complejos pueden aproximarse por distribuciones normales

**Pregunta central**: ¿Por qué la distribución normal es tan especial?

## Enunciado del Teorema del Límite Central

### Versión Simple

Sea $X_1, X_2, \ldots, X_n$ una secuencia de variables aleatorias **independientes e idénticamente distribuidas** (i.i.d.) con:
- Media: $\mu = \mathbb{E}[X_i]$
- Varianza: $\sigma^2 = \text{Var}(X_i) < \infty$

Definimos la **suma normalizada**:

$$Z_n = \frac{\sum_{i=1}^{n} X_i - n\mu}{\sigma\sqrt{n}} = \frac{S_n - n\mu}{\sigma\sqrt{n}}$$

Entonces:

$$\lim_{n \to \infty} P(Z_n \leq z) = \Phi(z) = \int_{-\infty}^{z} \frac{1}{\sqrt{2\pi}} e^{-x^2/2} dx$$

**En palabras**: La **distribución de la suma normalizada** converge a una **distribución normal estándar**, sin importar la distribución original.

### Implicación Práctica

La suma de muchas variables aleatorias sigue aproximadamente una **distribución normal**:

$$S_n = \sum_{i=1}^{n} X_i \approx \mathcal{N}(n\mu, n\sigma^2)$$

Para $n$ suficientemente grande (típicamente $n > 30$).

## Formulación Alternativa

Si definimos la **media muestral**:

$$\bar{X}_n = \frac{1}{n} \sum_{i=1}^{n} X_i$$

Entonces:

$$\sqrt{n}(\bar{X}_n - \mu) \xrightarrow{d} \mathcal{N}(0, \sigma^2)$$

O equivalentemente:

$$\bar{X}_n \approx \mathcal{N}\left(\mu, \frac{\sigma^2}{n}\right)$$

**Interpretación**: La media muestral es aproximadamente normal con varianza que decrece como $1/n$.

## ¿Por Qué es Especial?

### Universalidad

El TLC funciona para **casi cualquier distribución original**:
- Exponencial (colas pesadas)
- Uniforme (colas cortas)
- Bernoulli (discreta)
- Pareto (power-law)
- Cualquier otra distribución con varianza finita

### Velocidad de Convergencia

La convergencia es **sorprendentemente rápida**:
- Distribuciones simétricas: convergencia con $n \approx 20-30$
- Distribuciones asimétricas: convergencia con $n \approx 100-300$
- Distribuciones con colas muy pesadas: convergencia más lenta (requiere $n > 1000$)

### Robustez

El TLC es **robusto** a:
- Distribuciones no idénticas (si no son demasiado diferentes)
- Dependencia débil entre variables
- Outliers moderados

## Demostración Intuitiva

### Mecanismo de Convergencia

1. **Variabilidad se suaviza**: Cuando sumamos variables, los errores positivos y negativos se cancelan

2. **Central limit effect**: La distribución resultante se concentra alrededor de la media

3. **Forma gaussiana emerge**: Independientemente de la forma original

### Analogía Física

Como en el **movimiento browniano**: una partícula sufre innumerables choques microscópicos (variables aleatorias), y su posición final sigue una distribución normal.

## Aplicaciones en Econofísica

### 1. Retornos de Mercado

Los **retornos diarios** de un activo son la suma de muchos micro-eventos (transacciones, noticias, ajustes):

$$r_{\text{diario}} = \sum_{\text{todos los eventos}} r_{\text{evento}}$$

Por TLC: $r_{\text{diario}} \approx \mathcal{N}(\mu, \sigma^2)$

Esta es la base del **modelo de Black-Scholes**.

### 2. Agregación de Cambios

Los **cambios de precio agregados** sobre períodos más largos son más gaussianos:
- Cambios en 1 minuto: desviación de normalidad visible
- Cambios en 1 hora: más próximo a normal
- Cambios en 1 día: muy cercano a normal
- Cambios en 1 mes: casi perfectamente normal

### 3. Portafolios Diversificados

El retorno de un portafolio diversificado es más gaussiano que sus componentes:

$$r_{\text{portafolio}} = \sum_{i=1}^{N} w_i r_i$$

Cuantos más activos ($N$ grande), más gaussiano es el retorno del portafolio.

## Limitaciones del TLC

### 1. Requiere Varianza Finita

Para distribuciones con varianza infinita (ej: Pareto con $\alpha < 2$), el TLC **no funciona**.

En este caso, el límite es una **distribución estable de Lévy**, no gaussiana.

### 2. Convergencia Lenta para Colas Pesadas

Datos financieros con eventos extremos requieren $n$ muy grande:

$$n_{\text{requerido}} \propto \frac{1}{(\text{sesgo})^2}$$

### 3. Suposición de Independencia

Si las variables están altamente correlacionadas, el TLC se modifica (requiere teoremas más avanzados).

### 4. No Predice Colas Pesadas Observadas

El TLC predice distribución normal, pero datos reales tienen **colas más pesadas**.

Explicación: correlaciones dinámicas, efectos no lineales, cambios en régimen.

## Tabla Resumen

| Aspecto | Descripción |
|---|---|
| **Requisitos** | Variables i.i.d., varianza finita |
| **Conclusión** | Suma normalizada → Normal |
| **Velocidad** | 20-300 muestras típicamente |
| **Validez** | Casi universal |
| **Excepciones** | Colas muy pesadas (Lévy) |
| **Uso en finanzas** | Base de Black-Scholes, VaR, etc. |

## El Código: Demostración Visual

En este notebook demostraremos visualmente cómo el TLC funciona:

1. Tomamos una distribución **inicial arbitraria** (exponencial, Bernoulli, uniforme)
2. **Sumamos** muchas muestras de esa distribución
3. **Repetimos** miles de veces
4. Observamos que la **distribución resultante es gaussiana**

Este es uno de los experimentos más fundamentales en probabilidad.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# ============================================================================
# DEMOSTRACIÓN DEL TEOREMA DEL LÍMITE CENTRAL (TLC)
#
# Objetivo: Demostrar visualmente cómo la distribución de una SUMA
# de variables aleatorias converge a una distribución NORMAL,
# sin importar la distribución original.
# ============================================================================

def simular_tlc(dist_func, title_prefix, color='steelblue'):
    """
    Simula el Teorema del Límite Central con una distribución dada.
    
    PARÁMETROS CLAVE DEL TLC:
    ========================
    
    N_vars: Número de variables que sumamos
    - Esto es el "n" en el TLC
    - Mayor N_vars = mejor aproximación a normal
    - Típicamente: n > 30 es suficiente
    
    N_sims: Número de veces que repetimos el experimento
    - Cada simulación: suma N_vars variables aleatorias
    - Repetimos N_sims veces para estimar la distribución
    - Mayor N_sims = mejor estimación (más muestras)
    
    PROCESO:
    ========
    1. Generar una muestra de tamaño N_vars de la distribución original
    2. Calcular su SUMA
    3. Repetir N_sims veces
    4. Graficar la distribución de las sumas
    5. Comparar con distribución normal teórica
    """
    
    N_vars = 100        # Número de variables aleatorias que sumamos
                        # S_n = X_1 + X_2 + ... + X_100
                        # Por TLC: S_n ≈ Normal(100*μ, 100*σ²)
    
    N_sims = 10000      # Número de simulaciones (repeticiones del experimento)
                        # Cada simulación genera una suma
                        # 10,000 muestras de S_n para estimar su distribución

    print(f"\n{'='*70}")
    print(f"{title_prefix}: DEMOSTRACIÓN DEL TLC")
    print(f"{'='*70}")
    print(f"N_vars (cantidad de variables sumadas): {N_vars}")
    print(f"N_sims (número de simulaciones): {N_sims:,}")
    
    # GENERACIÓN DE SUMAS
    # ===================
    # Para cada simulación i:
    #   1. Generar N_vars números aleatorios de la distribución
    #   2. Sumarlos para obtener S_i
    # Resultado: array de 10,000 sumas
    
    sums = [np.sum(dist_func(N_vars)) for _ in range(N_sims)]
    
    # ANÁLISIS ESTADÍSTICO
    # ====================
    
    mu_sum = np.mean(sums)      # Media de las sumas
    sigma_sum = np.std(sums)    # Desv. estándar de las sumas
    
    print(f"\nEstadísticas de la distribución de sumas:")
    print(f"  Media (μ_sum): {mu_sum:.4f}")
    print(f"  Desv. Estándar (σ_sum): {sigma_sum:.4f}")
    print(f"  Mínimo: {np.min(sums):.4f}")
    print(f"  Máximo: {np.max(sums):.4f}")
    
    # VISUALIZACIÓN
    # =============
    
    plt.figure(figsize=(12, 6))
    
    # Histograma de las sumas observadas
    # Densidad = True para que el área total = 1 (probabilidad)
    counts, bins, patches = plt.hist(sums, bins=50, density=True, 
                                      alpha=0.7, edgecolor='black',
                                      color=color,
                                      label='Distribución de Sumas (Simulada)')
    
    # AJUSTE GAUSSIANO TEÓRICO
    # ========================
    # Según TLC: La distribución de sumas debe ser Normal(μ_sum, σ_sum²)
    # Calculamos la función de densidad gaussiana teórica
    
    x = np.linspace(np.min(sums), np.max(sums), 200)
    
    # Distribución normal con parámetros estimados
    normal_fit = norm.pdf(x, loc=mu_sum, scale=sigma_sum)
    
    plt.plot(x, normal_fit, 'r-', linewidth=3, 
             label=f'Distribución Normal Teórica (μ={mu_sum:.2f}, σ={sigma_sum:.2f})')
    
    # DECORACIÓN
    plt.title(f'{title_prefix}: {N_vars} Variables Sumadas\n' +
              f'Convergencia a Distribución Normal (N={N_sims:,} simulaciones)',
              fontsize=12, fontweight='bold')
    plt.xlabel('Suma de variables aleatorias', fontsize=11)
    plt.ylabel('Densidad de probabilidad', fontsize=11)
    plt.legend(fontsize=10, loc='upper right')
    plt.grid(True, alpha=0.3, linestyle='--')
    plt.tight_layout()
    plt.show()
    
    # VALIDACIÓN DEL TLC
    # ==================
    # Calculamos la "bondad de ajuste" usando test de normalidad
    
    from scipy.stats import shapiro
    
    # Test de Shapiro-Wilk (p-value > 0.05 sugiere distribución normal)
    _, p_value = shapiro(sums[:5000])  # Limitar a 5000 para eficiencia
    
    print(f"\nValidación de normalidad (Shapiro-Wilk test):")
    print(f"  p-value: {p_value:.6f}")
    if p_value > 0.05:
        print(f"  ✓ No hay evidencia significativa CONTRA normalidad")
    else:
        print(f"  ◐ Posible desviación de normalidad (puede ser por {N_vars} no grande)")
    
    print(f"\nCONCLUSIÓN: La suma de {N_vars} variables {title_prefix}")
    print(f"está bien descrita por una distribución Normal.")
    print(f"Esto es la esencia del Teorema del Límite Central.\n")

# ============================================================================
# APLICACIÓN 1: DISTRIBUCIÓN EXPONENCIAL
# ============================================================================
# La exponencial tiene COLAS PESADAS y es muy diferente de normal
# Parámetro: scale = 1 (media = 1)
# 
# PROPIEDAD: Los tiempos de espera siguen exponencial
# ECONOMÍA: Tiempos entre transacciones, duraciones de crisis
#
# PREGUNTA: ¿Qué pasa si sumamos 100 exponenciales?

print("\n" + "#"*70)
print("# APLICACIÓN 1: DISTRIBUCIÓN EXPONENCIAL")
print("#"*70)
print("""
La distribución exponencial modela:
- Tiempos de espera entre eventos
- Duración de ciclos económicos
- Tiempo entre transacciones en mercados

Características:
- Altamente asimétrica (colas largas a la derecha)
- NO es similar a normal
- Parámetro: tasa de decaimiento

Pregunta: ¿Converge a normal cuando sumamos 100 exponenciales?
""")

simular_tlc(lambda N: np.random.exponential(size=N), 
            'Exponencial', 
            color='coral')

# ============================================================================
# APLICACIÓN 2: DISTRIBUCIÓN DE BERNOULLI
# ============================================================================
# Bernoulli es DISCRETA (solo toma valores 0 o 1)
# p = 0.5 significa 50% de probabilidad para cada valor
#
# PROPIEDAD: Resultados de eventos binarios (éxito/fracaso)
# ECONOMÍA: Mercados sube/baja, predicciones binarias
#
# PREGUNTA: ¿Qué pasa si sumamos 100 Bernoulli?

print("\n" + "#"*70)
print("# APLICACIÓN 2: DISTRIBUCIÓN DE BERNOULLI")
print("#"*70)
print("""
La distribución Bernoulli modela:
- Resultados binarios (éxito/fracaso, sí/no)
- Lanzamientos de moneda
- Decisiones de compra/venta en mercados

Características:
- Completamente discreta (solo 0 o 1)
- MUY diferente a normal
- NADA similar en apariencia

Pregunta: ¿Converge a normal cuando sumamos 100 Bernoulli?
La suma de 100 Bernoulli es una distribución binomial.
Con n=100, p=0.5: debe ser aproximadamente normal.
""")

simular_tlc(lambda N: np.random.choice([0, 1], size=N), 
            'Bernoulli (Binomial)', 
            color='mediumseagreen')

# ============================================================================
# APLICACIÓN 3: DISTRIBUCIÓN UNIFORME
# ============================================================================
# Uniforme en [0,1]: todos los valores tienen igual probabilidad
#
# PROPIEDAD: Generador de números aleatorios base
# ECONOMÍA: Modelar incertidumbre sin información
#
# PREGUNTA: ¿Qué pasa si sumamos 100 uniformes?

print("\n" + "#"*70)
print("# APLICACIÓN 3: DISTRIBUCIÓN UNIFORME")
print("#"*70)
print("""
La distribución uniforme modela:
- Incertidumbre cuando toda información es igual de probable
- Base para generar otras distribuciones
- Escenarios sin sesgos conocidos

Características:
- Distribución rectangular/plana
- Iguales probabilidades en todo el rango
- Ni colas pesadas, ni asimétricas

Pregunta: ¿Converge a normal cuando sumamos 100 uniformes?
""")

simular_tlc(lambda N: np.random.rand(N), 
            'Uniforme', 
            color='mediumpurple')

# ============================================================================
# CONCLUSIONES Y SÍNTESIS
# ============================================================================

print("\n" + "="*70)
print("CONCLUSIONES DEL TEOREMA DEL LÍMITE CENTRAL")
print("="*70)
print("""
KEY FINDINGS:
=============

1. UNIVERSALIDAD:
   ✓ Comenzamos con 3 distribuciones COMPLETAMENTE diferentes:
     - Exponencial: colas pesadas, asimétricas
     - Bernoulli: discreta, solo dos valores
     - Uniforme: simétrica, sin preferencias
   
   ✓ Todas convergen a la MISMA distribución: NORMAL

2. MECANISMO:
   Cuando sumamos N variables aleatorias:
   - Los errores positivos y negativos se promedian
   - La variabilidad total disminuye
   - La forma de la distribución se "suaviza"
   - Emerge una forma gaussiana

3. VELOCIDAD:
   Con N = 100 variables:
   - Ya tenemos aproximación excelente a normal
   - N >> 30 es generalmente suficiente
   - Distribuciones simétricas convergen rápido

4. APLICACIONES EN ECONOFÍSICA:
   
   a) RETORNOS DE MERCADO:
      - Retorno diario = suma de micro-eventos
      - Por TLC: aproximadamente normal
      - Base del modelo Black-Scholes
   
   b) PORTAFOLIOS:
      - Retorno = suma ponderada de activos
      - Mas activos = mas gaussiano
      - Diversificación → normalidad
   
   c) AGREGACIÓN TEMPORAL:
      - Retornos por minuto: NO normal (colas pesadas)
      - Retornos por hora: MÁS normal
      - Retornos por día: CASI normal
      - Retornos por mes: MUY normal

5. LIMITACIONES:
   - Requiere varianza finita (excluye Cauchy, Pareto severo)
   - Requiere variables independientes
   - Convergencia lenta si distribución original muy asimétrica
   - Datos reales pueden tener desviaciones (colas más pesadas)

6. POR QUÉ IMPORTA:
   El TLC es la razón por la que la distribución normal:
   - Aparece en tantos fenómenos naturales
   - Es tan útil en estadística
   - Fundamenta la mayor parte de la econometría
   - Justifica modelos como Black-Scholes

IMPLICACIÓN PROFUNDA:
====================
La distribución normal no es "elegida" por la naturaleza
porque sea bonita o simple. Es que EMERGE necesariamente
de cualquier proceso que agregue muchas variables aleatorias.

Es casi una ley universal de la probabilidad.
""")

print("="*70)
```
