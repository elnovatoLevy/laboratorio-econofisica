# Momentos Estadísticos en Distribuciones Financieras

## Contexto: ¿Por qué importan los momentos en econofísica?

En finanzas, es común asumir que los retornos de activos siguen una **distribución normal** (Gaussiana). Esta suposición simplifica muchos modelos (ej: Black-Scholes para opciones).

**Problema**: Los datos reales no son normales. Observamos:
- **Colas pesadas** (eventos extremos más frecuentes)
- **Asimetría** (caídas más abruptas que subidas)

Los **momentos estadísticos** caracterizan estas desviaciones de la normalidad y son cruciales para:
- Gestión de riesgo (Value at Risk - VaR)
- Valoración de derivados
- Detección de anomalías de mercado

## Los Cuatro Momentos Estadísticos

### 1. **Media (Primer Momento)**

$$\mu = \langle x \rangle = \frac{1}{N} \sum_{i=1}^{N} x_i$$

**Significado**: Tendencia central, retorno promedio esperado.

**En finanzas**:
- Media positiva → Activo con tendencia alcista
- Media negativa → Activo con tendencia bajista
- Media ≈ 0 → Activo neutral (caso común en retornos logarítmicos)

### 2. **Varianza y Desviación Estándar (Segundo Momento Central)**

$$\sigma^2 = \langle (x - \mu)^2 \rangle = \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2$$

$$\sigma = \sqrt{\sigma^2}$$

**Significado**: Dispersión alrededor de la media, volatilidad.

**En finanzas**:
- $\sigma$ **bajo** → Activo estable, bajo riesgo
- $\sigma$ **alto** → Activo volátil, alto riesgo
- Métrica clave en modelo CAPM: riesgo = desviación estándar

### 3. **Asimetría / Skewness (Tercer Momento Central)**

$$\gamma_1 = \frac{\langle (x - \mu)^3 \rangle}{\sigma^3}$$

**Interpretación**:

$$\gamma_1 = \begin{cases}
\gamma_1 > 0 & \text{Sesgo positivo (cola derecha larga)} \\
\gamma_1 = 0 & \text{Distribución simétrica} \\
\gamma_1 < 0 & \text{Sesgo negativo (cola izquierda larga)}
\end{cases}$$

**Visualización**:
- $\gamma_1 > 0$: Distribución con cola extendida a la derecha (eventos positivos raros pero posibles)
- $\gamma_1 < 0$: Distribución con cola extendida a la izquierda (eventos negativos raros pero posibles)
- $\gamma_1 \approx 0$: Distribución simétrica (como la normal)

**Significado financiero**:
- Distribuciones de retornos típicamente tienen $\gamma_1 < 0$ (asimetría negativa)
- Significa: caídas bruscas ocasionales (crash risk) son más probables que subidas bruscas
- Implica: invertores requieren **prima de riesgo** adicional

### 4. **Curtosis / Kurtosis (Cuarto Momento Central)**

$$\gamma_2 = \frac{\langle (x - \mu)^4 \rangle}{\sigma^4} - 3$$

La resta de 3 hace que la **curtosis en exceso** sea 0 para la distribución normal.

**Interpretación**:

$$\gamma_2 = \begin{cases}
\gamma_2 > 0 & \text{Leptocúrtica (colas pesadas, pico agudo)} \\
\gamma_2 = 0 & \text{Mesocúrtica (normal)} \\
\gamma_2 < 0 & \text{Platicúrtica (colas ligeras, pico suave)}
\end{cases}$$

**Visualización**:
- $\gamma_2 > 0$: Pico alto en el centro + colas gruesas (eventos extremos más probables)
- $\gamma_2 = 0$: Distribución normal estándar
- $\gamma_2 < 0$: Pico bajo + colas finas (eventos extremos menos probables)

**Significado financiero**:
- Datos reales suelen tener $\gamma_2 > 0$ (colas pesadas)
- Significa: crashes y rallies son más frecuentes que predice la teoría normal
- Implicación: Modelos de riesgo basados en distribución normal **subestiman** riesgo de tail
- Ejemplo: Basilea III requiere capital adicional por "riesgo de cola"

## Ejemplos Numéricos

| Distribución | Media | σ | Skewness | Excess Kurtosis |
|---|---|---|---|---|
| Normal | 0 | 1 | 0 | 0 |
| t-Student (df=3) | 0 | ∞ | 0 | ∞ |
| Lognormal | e^(μ+σ²/2) | - | > 0 | > 0 |
| Retornos reales (stocks) | ~0 | 0.01-0.02 | -0.2 a -0.5 | 3-10 |

## Distribución t-Student vs Normal

La **distribución t-Student** es una alternativa a la normal que:
- Tiene **colas más pesadas** (parámetro df controla qué tan pesadas)
- Con **df pequeño** (ej: 3-5), modela bien eventos extremos reales
- Es más realista para retornos financieros

$$P(x | \nu) = \frac{\Gamma((\nu+1)/2)}{\sqrt{\pi \nu} \Gamma(\nu/2)} \left(1 + \frac{x^2}{\nu}\right)^{-(\nu+1)/2}$$

Donde $\nu$ (df) es los grados de libertad.

## Implicaciones para Gestión de Riesgo

1. **Modelos basados en Normal pueden ser peligrosos**
   - Subestiman probabilidad de eventos extremos
   - "Black swans" son más comunes de lo predicho

2. **Necesidad de medidas de riesgo alternativas**
   - Value at Risk (VaR)
   - Expected Shortfall (ES)
   - Stress testing

3. **Aplicaciones prácticas**
   - Requisitos de capital bancario
   - Hedge de opciones
   - Diversificación de portafolio

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, norm

# ============================================================================
# ANÁLISIS DE MOMENTOS ESTADÍSTICOS EN DISTRIBUCIONES FINANCIERAS
# ============================================================================

np.random.seed(42)
n_observations = 1260

# Generar retornos con distribución t-Student (colas pesadas)
returns = np.random.standard_t(df=10, size=n_observations) * 0.01 + 0.0005

# Agregar eventos extremos
extreme_indices = np.random.choice(n_observations, size=10, replace=False)
returns[extreme_indices] *= 3

# Visualización
plt.figure(figsize=(12, 6))
plt.hist(returns, bins=100, density=True, alpha=0.7, edgecolor='black',
         color='steelblue', label='Distribución empírica')

x = np.linspace(returns.min(), returns.max(), 100)
plt.plot(x, norm.pdf(x, loc=np.mean(returns), scale=np.std(returns)),
         'r--', linewidth=2, label='Distribución Normal (comparación)')

plt.title('Distribución de Retornos: Simulado vs Normal\nNótese colas más pesadas',
          fontsize=13, fontweight='bold')
plt.xlabel('Retorno (log-return)', fontsize=11)
plt.ylabel('Densidad', fontsize=11)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()

# ============================================================================
# MOMENTOS ESTADÍSTICOS
# ============================================================================

print("="*70)
print("ANÁLISIS DE MOMENTOS ESTADÍSTICOS")
print("="*70)

# 1. ASIMETRÍA (SKEWNESS)
asimetria = skew(returns)

print(f"\n1. ASIMETRÍA (Skewness) = {asimetria:.4f}")
print(f"   Mide: Simetría de la distribución")
print(f"   > 0: Cola derecha larga (sesgo positivo)")
print(f"   = 0: Distribución simétrica (como la normal)")
print(f"   < 0: Cola izquierda larga (crash risk)")

if abs(asimetria) > 0.5:
    print(f"   ✓ Asimetría importante detectada")
    if asimetria < 0:
        print(f"   ✗ Sesgo NEGATIVO: Crash risk probable")
    else:
        print(f"   ✓ Sesgo POSITIVO: Upside potential")
elif abs(asimetria) > 0.05:
    print(f"   ◐ Asimetría moderada")
else:
    print(f"   ◯ Distribución aproximadamente simétrica")

# 2. CURTOSIS (KURTOSIS)
exceso_curtosis = kurtosis(returns)
curtosis_total = exceso_curtosis + 3

print(f"\n2. CURTOSIS")
print(f"   Exceso Curtosis = {exceso_curtosis:.4f}")
print(f"   Curtosis Total = {curtosis_total:.4f} (Normal = 3)")
print(f"\n   Mide: Pesadez de colas")
print(f"   > 0: Colas pesadas (Leptocúrtica) - Black swans")
print(f"   = 0: Colas normales")
print(f"   < 0: Colas ligeras (Platicúrtica)")

if exceso_curtosis > 1:
    print(f"   ✗ COLAS PESADAS: Eventos extremos frecuentes")
    print(f"   → Modelos normales subestiman riesgo")
elif exceso_curtosis > -1:
    print(f"   ◯ COLAS NORMALES: Similar a distribución normal")
else:
    print(f"   ◐ COLAS LIGERAS: Menos eventos extremos")

# ============================================================================
# CONCLUSIÓN
# ============================================================================

print(f"\n" + "="*70)
print(f"CONCLUSIÓN")
print(f"="*70)

if exceso_curtosis > 0:
    print(f"\nLos retornos presentan colas pesadas (leptocúrticos)")
    print(f"Indica: Mayor probabilidad de eventos extremos que la normal")
else:
    print(f"\nLos retornos son platicúrticos (colas más ligeras)")

if abs(asimetria) > 0.05:
    print(f"La distribución está sesgada")
    print(f"Implica: Asymmetric risk profile para inversores")
else:
    print(f"La distribución es aproximadamente simétrica")

print(f"\n" + "="*70)
```
